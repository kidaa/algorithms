````python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
````

### Exercises 1 
```
1. Write code necessary to analyze the relationship between median income and recycling rate in New York City Community Boards (using 2013_NYC_CD_MedianIncome_Recycle.xlsx). Calculate:
• coefficient of correlation
• coefficient of determination
```

````python
df = pd.read_excel("data/2013_NYC_CD_MedianIncome_Recycle.xlsx")
df
````

````python
df.plot(kind="scatter",x="MdHHIncE",y="RecycleRate") # do a scatter plot of the data
````

````python
# calculate the coefficient of correlation
df.corr()
````

````python
data = np.asarray(df[['RecycleRate','MdHHIncE']]) #convert the dataframe to a nparray
#data
````

````python
x, y = data[:, 1:], data[:, 0] #assign the values to variables and reshape the data
#data[:, 1:]
#data[:, 0]
#data
````

````python
lr = LinearRegression() # create a LinearRegression object
````

````python
lr.fit(x,y) #fit the data to the values
````

````python
m = lr.coef_[0] # the slope of our regression line
m
````

````python
b = lr.intercept_ # the y-intercept (value when x = 0)
b
````

````python
# plot the linear regression line on the scatter plot
df.plot(kind="scatter",x="MdHHIncE",y="RecycleRate")
plt.plot(df['MdHHIncE'],m*df['MdHHIncE']+b,'-')
````

````python
# calculate the R-squared (coefficient of determination)
lr.score(x,y)
````
